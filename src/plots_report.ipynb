```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial import ConvexHull
```

#Simplex

```python

pts = np.array([[1, 3, 5], [4, 2, 1], [2, 2, 3], [0, 1, 0], ])

hull = ConvexHull(pts)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(pts.T[0], pts.T[1], pts.T[2], "ko")
ax.axis('off')
for s in hull.simplices:
    s = np.append(s, s[0])
    ax.plot(pts[s, 0], pts[s, 1], pts[s, 2], "b-")

plt.show()
```

#Torus

```python
n = 100

theta = np.linspace(0, 3.*np.pi, n)
phi = np.linspace(0, 2*np.pi, n)
theta, phi = np.meshgrid(theta, phi)
R, r = 3, 1
x = (R + r*np.cos(theta)) * np.cos(phi)
y = (R + r*np.cos(theta)) * np.sin(phi)
z = r * np.sin(theta)

fig = plt.figure(figsize=(22, 20),)
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-3,3)
ax1.plot_surface(x, y, z, rstride=5, cstride=5, color='red', edgecolors='orange')
ax1.view_init(36, 26)
ax1.axis('off')
plt.show()
```

#Sampling uniformly on a Torus

```python
R, r = 3, 1

N = 5*10**3
simu_x = []
simu_y = []
simu_z = []

for i in range(N):
  U_x = 0
  U_y = 0
  U_z = r+1

  while (R-np.sqrt(U_x**2+U_y**2))**2 + U_z**2 > r**2:
    U_x = 7*np.random.rand()-4
    U_y = 8*np.random.rand()-4
    U_z = 2*np.random.rand()-1
  simu_x.append(U_x)
  simu_y.append(U_y)
  simu_z.append(U_z)

fig = plt.figure(figsize=(22, 20),)
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-3,3)
ax1.scatter(simu_x,simu_y,simu_z)
plt.show()
```

#Uniform sample on double torus

```python
r = 0.01

N = 6*10**3
simu_x = []
simu_y = []
simu_z = []

for i in range(N):
  U_x = 0
  U_y = 0
  U_z = r+1

  while (U_x*(U_x-1)**2*(U_x-2) + (U_y/5)**2)**2 + (U_z/10)**2 > r:
    U_x = 4*np.random.rand()-2
    U_y = 8*np.random.rand()-4
    U_z = 4*np.random.rand()-2
  simu_x.append(U_x)
  simu_y.append(U_y)
  simu_z.append(U_z)

fig = plt.figure(figsize=(22, 20),)
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-3,3)
ax1.scatter(simu_x,simu_y,simu_z)
plt.show()
```

#Sampling from double torus with additional noise

```python
r = 0.01

N = 6*10**3
simu_x = []
simu_y = []
simu_z = []

for i in range(N):
  U_x = 0
  U_y = 0
  U_z = r+1

  while (U_x*(U_x-1)**2*(U_x-2) + (U_y/5)**2)**2 + (U_z/10)**2 > r:
    U_x = 4*np.random.rand()-2
    U_y = 8*np.random.rand()-4
    U_z = 4*np.random.rand()-2
  simu_x.append(U_x + np.random.normal(0,0.065))
  simu_y.append(U_y+np.random.normal(0,0.065))
  simu_z.append(U_z+np.random.normal(0,0.065))

fig = plt.figure(figsize=(22, 20),)
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-3,3)
ax1.scatter(simu_x,simu_y,simu_z)
plt.show()
```
